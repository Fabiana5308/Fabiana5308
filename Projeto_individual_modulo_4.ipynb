#<font color = gren> Relatórios de Progresso

#Introdução
Este relatório tem como objetivo apresentar o progresso diário do projeto de desenvolvimento de software ao longo de uma semana. Serão analisados diversos indicadores para avaliar a produtividade da equipe, incluindo o total e média de horas trabalhadas, total e média de bugs corrigidos, total e média de tarefas concluídas e a produtividade diária medida pelo número de tarefas concluídas por hora.

```python
#importando as bibliotecas pra análise e para os gráficos

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dados fornecidos
informacao = {
    'Dias': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado', 'Domingo'],
    'Horas Trabalhadas': [8, 7, 6, 8, 8, 6, 5],
    'Bugs Corrigidos': [3, 3, 4, 5, 2, 3, 2],
    'Tarefas Concluídas': [6, 5, 5, 6, 4, 3, 3]
}
```

```python
#criando DataFrame

df = pd.DataFrame(informacao)
df
```

```python
df['Horas Trabalhadas'].sum()
```

```python
media_horas_trabalhadas = round(df['Horas Trabalhadas'].mean(), 2)
print("Média Diária de Horas Trabalhadas:", media_horas_trabalhadas, "horas")
```

```python
# Criação do DataFrame
df = pd.DataFrame(informacao)

# Cálculo das métricas
total_horas_trabalhadas = df['Horas Trabalhadas'].sum()
media_horas_trabalhadas = round(df['Horas Trabalhadas'].mean(), 2)

# Plot do gráfico de barras de comparação para total e média de horas trabalhadas
plt.figure(figsize=(10, 8))
plt.ylim(0, 10,8)
bars = plt.bar(df['Dias'], df['Horas Trabalhadas'], color='blue', label='Horas Trabalhadas')
plt.axhline(y=media_horas_trabalhadas, color='orange', linestyle='dashed', label='Média')

# Adicionar os valores acima das barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(height), ha='center', va='bottom', fontsize=10)

plt.ylabel('Horas Trabalhadas')
plt.title('Total e Média de Horas Trabalhadas por Dia da Semana')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

#Total de Horas Trabalhadas:
* O projeto teve um total de 48 horas trabalhadas durante a semana. Isso indica que houve dedicação suficiente dos funcionários para o desenvolvimento do projeto.

#Média Diária de Horas Trabalhadas:
* A média diária de horas trabalhadas foi de aproximadamente 6.86 horas. Isso pode indicar uma distribuição equilibrada do trabalho ao longo da semana, com uma carga horária adequada por dia.



```python
df['Bugs Corrigidos'].sum()
```

```python
media_bugs_corrigidos = round(df['Bugs Corrigidos'].mean(), 2)
print("Média Diária de Bugs Corrigidos:", media_bugs_corrigidos, "bugs")
```

```python
# Criação do DataFrame
df = pd.DataFrame(informacao)

# Cálculo das métricas
total_bugs_corrigidos = df['Bugs Corrigidos'].sum()
media_bugs_corrigidos = round(df['Bugs Corrigidos'].mean(), 2)

# Plot do gráfico de barras de comparação para total e média de bugs corrigidos
plt.figure(figsize=(6, 3))
plt.ylim(0, 6,3)
bars = plt.bar(df['Dias'], df['Bugs Corrigidos'], color='green', label='Bugs Corrigidos')
plt.axhline(y=media_bugs_corrigidos, color='orange', linestyle='dashed', label='Média')

# Adicionar os valores acima das barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(height), ha='center', va='bottom', fontsize=10)

plt.ylabel('Bugs Corrigidos')
plt.title('Total e Média de Bugs Corrigidos por Dia da Semana')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

#Total de Bugs Corrigidos:
* Foram corrigidos 22 bugs ao longo da semana. É importante monitorar esse número, pois bugs podem impactar na qualidade e no prazo do projeto.

#Média Diária de Bugs Corrigidos:
* A média diária de bugs corrigidos foi de aproximadamente 3.14 bugs. Isso pode sugerir que a equipe está conseguindo gerenciar bem as correções de bugs no decorrer dos dias.

```python
df['Tarefas Concluídas'].sum()
```

```python
media_tarefas_concluidas = round(df['Tarefas Concluídas'].mean(), 2)
print("Média Diária de Tarefas Concluídas:", media_tarefas_concluidas, "tarefas")
```

```python
# Criação do DataFrame
df = pd.DataFrame(informacao)

# Cálculo das métricas
total_tarefas_concluidas = df['Tarefas Concluídas'].sum()
media_tarefas_concluidas = round(df['Tarefas Concluídas'].mean(), 2)

# Plot do gráfico de barras de comparação para total e média de Tarefas Concluídas
plt.figure(figsize=(8, 4))
plt.ylim(0, 8,4)
bars = plt.bar(df['Dias'], df['Tarefas Concluídas'], color='green', label='Tarefas Concluídas')
plt.axhline(y=media_tarefas_concluidas, color='orange', linestyle='dashed', label='Média')

# Adicionar os valores acima das barras
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(height), ha='center', va='bottom', fontsize=8)

plt.ylabel('Tarefas Concluídas')
plt.title('Total e Média de Tarefas Concluídas por Dia da Semana')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

#Total de Tarefas Concluídas:
* Um total de 32 tarefas foram concluídas durante a semana. É relevante acompanhar esse indicador para verificar o progresso do projeto.

#Média Diária de Tarefas Concluídas:
* A média diária de tarefas concluídas foi de aproximadamente 4.57 tarefas. Isso pode indicar uma boa eficiência da equipe na execução das atividades.

```python
# Cálculo da produtividade diária
df['Produtividade Diária'] = df['Tarefas Concluídas'] / df['Horas Trabalhadas']
df['Produtividade Diária'] = df['Produtividade Diária'].apply(lambda x: round(x, 2))
print("Produtividade Diária (Tarefas Concluídas por Hora):")
for i, row in df.iterrows():
    print(f"{row['Produtividade Diária']} tarefas por hora na {row['Dias']}-feira")
```

#Produtividade Diária:
* A produtividade diária, medida em tarefas concluídas por hora, variou ao longo da semana.

* **Algumas sugestões para análise são:**
Às quartas-feiras, a produtividade foi a mais alta (0.83 tarefas por hora), o que pode indicar maior foco e eficiência nesse dia.
Às sextas-feiras e sábados, a produtividade foi a mais baixa (0.5 tarefas por hora), talvez seja interessante analisar se esses dias estão sendo mais desafiadores para a equipe ou se existem fatores externos que influenciam.

```python
# Exibição do relatório
print("Relatório de Progresso Diário - Projeto de Desenvolvimento de Software")
print("\nTotal de Horas Trabalhadas:", total_horas_trabalhadas, "horas")
print("Média Diária de Horas Trabalhadas:", media_horas_trabalhadas, "horas")
print("Total de Bugs Corrigidos:", total_bugs_corrigidos, "bugs")
print("Média Diária de Bugs Corrigidos:", media_bugs_corrigidos, "bugs")
print("Total de Tarefas Concluídas:", total_tarefas_concluidas, "tarefas")
print("Média Diária de Tarefas Concluídas:", media_tarefas_concluidas, "tarefas")
print("Produtividade Diária (Tarefas Concluídas por Hora):")
for i, row in df.iterrows():
    print(f"{row['Produtividade Diária']} tarefas por hora na {row['Dias']}-feira")
```

```python
# Criação do DataFrame
df = pd.DataFrame(informacao)

# Unir os dados em um formato "tidy" para o Seaborn

df_melted = df.melt(id_vars='Dias', var_name='análise', value_name='Quantidade')

# Configuração do tamanho do gráfico
plt.figure(figsize=(10, 5))
plt.ylim(0, 10,5)

# Plot do gráfico de barras lado a lado para comparar as análises diárias
sns.barplot(x='Dias', y='Quantidade', hue='análise', data=df_melted, palette='viridis')

# Definindo os rótulos dos eixos e título do gráfico

plt.ylabel('Quantidade')
plt.title('análises Diárias do Projeto')

# Rotacionar os rótulos do eixo x para facilitar a leitura
plt.xticks(rotation=45)

# Exibir a legenda
plt.legend()

# Exibir o gráfico
plt.show()
```
